### Составление справочника населенных пунктов
загрузка справочника населенных пунктов JSON с hh.ru

```python
#создадим справочник country_dict (с информацией о странах, регионах и населенных пунктах) на основе данных с hh.ru
# Чтение данных из JSON-файла
with open('areas hh.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Функция для обработки одной страны
def process_country(country):
    # Верхний уровень (страна)
    country_top = pd.json_normalize(country)[['id', 'parent_id', 'name']]

    # Средний уровень (области)
    middle_level = pd.json_normalize(country['areas'], record_path=None, meta=['id', 'parent_id', 'name'])
    middle_level = middle_level.add_prefix('_1')

    # Нижний уровень (города)
    third_level = []
    for region in country['areas']:
        if len(region.get('areas', [])) > 0:
            city_data = pd.json_normalize(region['areas']).add_prefix('_2')
            city_data['_region_id'] = region['id']
            third_level.append(city_data)
            
    # Если есть третий уровень, собираем его
    if third_level:
        third_frame = pd.concat(third_level, ignore_index=True)
        combined_middle_and_third = pd.merge(
            middle_level,
            third_frame,
            left_on="_1id",
            right_on="_region_id",
            how="left"
        ).drop(columns="_region_id")
    else:
        combined_middle_and_third = middle_level

    # Объединяем верхний и средний уровни
    country_dict = pd.merge(
        country_top.assign(key=1),
        combined_middle_and_third.assign(key=1),
        on="key",
        how="outer"
    ).drop(columns="key")

    country_dict = country_dict.rename(columns={
        '_1id': 'id_1',
        '_1parent_id': 'parent_id_1',
        '_1name': 'name_1',
        '_2id': 'id_2',
        '_2parent_id': 'parent_id_2',
        '_2name': 'name_2'
    })

    return country_dict

# Обрабатываем каждую страну отдельно
all_dfs = []
for country in data:
    result = process_country(country)
    all_dfs.append(result)

# Объединяем все обработанные страны
country_dict = pd.concat(all_dfs, ignore_index=True)
country_dict = country_dict[['id', 'parent_id', 'name', 'id_1', 'parent_id_1', 'name_1', 'id_2', 'parent_id_2', 'name_2']]

print(country_dict.shape)
country_dict.tail(6)
```

```python
country_dict.head(6)
```

```python
# сделаем справочник населенных пунктов из РФ
RFcountry_dict = country_dict[country_dict['name'] == 'Россия']
RFcountry = pd.concat([
    RFcountry_dict[['id', 'parent_id',  'name']],         
    RFcountry_dict[['id_1', 'parent_id_1', 'name_1']].rename(columns={'id_1':'id', 'parent_id_1': 'parent_id','name_1':'name'}), 
    RFcountry_dict[['id_2', 'parent_id_2', 'name_2']].rename(columns={'id_2':'id', 'parent_id_2': 'parent_id','name_2':'name'})   
]).reset_index(drop=True)

#далее удаляем рустые значения id (пришли из пустых значений id_2)
RFcountry = RFcountry.dropna(subset=['id'])
#удаляем дубликаты
RFcountry = RFcountry.drop_duplicates(subset=["id"])
RFcountry.rename(columns={'name': 'area_RF_name'})
print(RFcountry.shape)
RFcountry
```
